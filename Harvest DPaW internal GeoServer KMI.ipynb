# Harvesting KMI GeoServer 2.8.0
http://internal-data.dpaw.wa.gov.au/ harvests http://kmi.dpaw.wa.gov.au/, a GeoServer v2.8.0.
This will only work with DPaW credentials (for kmi) and from within the DPaW intranet (to access internal-data).

A file `sources.py` needs to contain the CKAN and SOURCE config as follows:

```
CKAN = {
  "dpaw-internal":{
    "url": "http://internal-data.dpaw.wa.gov.au/",
    "key": "API-KEY" 
  }
}

SOURCES = {    
  "kmi": {
    "url": "http://kmi.dpaw.wa.gov.au/geoserver/ows?service=wms&request=GetCapabilities",
    "un": "USERNAME",
    "pw": "PASSWORD"
  }
}
```

## Configure CKAN and source

````python
import ckanapi
from harvest_helpers import *
from secret import CKAN, SOURCES

ckan = ckanapi.RemoteCKAN(CKAN["dpaw-internal"]["url"], apikey=CKAN["dpaw-internal"]["key"])
print("Using CKAN {0}".format(ckan.address))
````

## Harvest

````python
kmi = WebMapService(SOURCES["kmi"]["url"], parse_remote_metadata=True,
                    username=SOURCES["kmi"]["un"], password=SOURCES["kmi"]["pw"])

# These orgs should exist in your CKAN with the following slugs:
#org_names = set([x.split(":")[0] for x in list(kmi.contents)])

# l_kmi will be a list of CKAN package_show-like dicts to upsert datasets from
l_kmi = get_layer_dict_gs28(kmi, SOURCES["kmi"]["url"], ckan, res_format="WMS", debug=True)
````

````python
# p_kmi will be a dict of package_show for all created/updated packages
p_kmi = upsert_datasets(l_kmi, ckan, overwrite_metadata=True, drop_existing_resources=True)

print("{0} datasets created or updated from {1} KMI WMS layers".format(len(p_kmi), len(kmi.contents)))
````
